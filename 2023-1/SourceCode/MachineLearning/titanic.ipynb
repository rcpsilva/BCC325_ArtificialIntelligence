```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rcpsilva/BCC325_ArtificialIntelligence/main/2023-1/SourceCode/MachineLearning/train.csv')
```

```python
df.head(2)
```

```python
att = ['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
df = df[att]
```

```python
df.head(2)
```

```python
df.describe()
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'])
```

```python
!pip install scikit-learn
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer()
idf=pd.DataFrame(imputer.fit_transform(df))
idf.columns=df.columns
idf.index=df.index
```

```python
idf.head(20)
```

```python
idf.describe()
```

```python
plt.hist(idf['Age'])
plt.hist(df['Age'])
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dt = DecisionTreeClassifier(max_depth=3)

X = idf[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = idf['Survived']
```

```python
dt.fit(X,y)
```

```python
fig = plt.figure(figsize=(25,20))
_ = plot_tree(dt, 
                   feature_names=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'],  
                   class_names=['unlucky','survivor'],
                   filled=True)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
scores = cross_val_score(dt, X, y, cv=10)
scores
```

```python
print(f'DT: {np.mean(scores)} +- {np.std(scores)}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=3)
lr = LogisticRegression(max_iter=500, penalty=None)
ann = MLPClassifier(hidden_layer_sizes=[5,5,5])
rf = RandomForestClassifier(max_depth=3) 

models = [dt,lr,ann,rf]
```

```python
scores = []
for m in models:
    scores.append(cross_val_score(m, X, y, cv=10))
```

```python
for i in range(len(models)):
    print(f'{models[i]}: \t{np.mean(scores[i]):.2f} +- {np.std(scores[i]):.2f}')
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10,100,500,1000], 
              'max_depth':[2, 3, 4, None],
              'max_features':['sqrt','log2',None]}

clf = GridSearchCV(rf, parameters)
clf.fit(X,y)
```
